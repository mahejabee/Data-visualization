#### 1. create a bar plot for count the frequency of the three species.

```python
# pip install matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python

sns.countplot(data["Species"])
```

```python
#There are around 50 setosa,50 versicolor and 50 virginica  flowers.
#There are equal number of setosa,versicolor and virginica flowers.
```

##### 2. Create a pie chart to show the percentage of species.

```python
plt.pie([50,50,50],labels=["iris-setosa","iris-versicolor","iris-verginica"],autopct="%0.0f%%",explode=[0,0,0],radius=2)
plt.title("species")
plt.show()
```

```python
#There are 33% of setosa flowers,33% of verginica flowers and 33% of versicolor flowers
```

##### 3. Create the joint to all numberical variable for analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
sns.jointplot(data=data,x="SepalLengthCm",y="PetalLengthCm",hue="Species")
```

```python
# setosa flowers have petal length of 1cm-2cm and sepal length of 4cm-6cm.
# versicolor flowers have petal length of 3cm-5cm and sepal length of 5cm-7.5cm.
# verginica flowers have petal length of 5cm-7cm and sepal length of 5cm-8cm.
# overall we can infer that setosa flowers are smaller in terms of petal length and sepal length when compared to versicolor and verginica 
```

##### 4. Create a scatter plot and take hue = "Species"

```python
m=sns.FacetGrid(data=data,hue="Species")
m.map(plt.scatter,"PetalLengthCm","PetalWidthCm")
m.add_legend()
```

```python
# setosa flowers are smaller than versicolor and verginica in terms of petal length and petal width.
```

```python
# All flowers with petal length 0-2cm and petal width 0-0.8cm are setosa and petal length 3-5cm and petal width of 1-1.8cm are versicolor and petal length of 4.2-6.5cm are virginica
```

###### 5. Create a Boxplot or Whisker plot as univeriate and bivariate analysis.

```python
sns.boxplot(data=data,y="PetalLengthCm",x="Species",hue="Species")
```

```python
# There are outliers in setosa flower.
# median value for setosa flower is 1.5 cm
# minimum petal length value is 1.2cm and max value is 1.8cm.
# There are very few outliers in versicolor.
# minimum petal length value for versicolor is 3.2cm and max is 5cm.
# The avg petal length value for versicolor is 4.5cm.
# There are no outliers in virginica flower.
# The avg petal length value for virginica is 5.8cm and min value is 4.8cmand max value is 7cm.
```

###### 6. Create a Strip plot 

```python
sns.stripplot(data=data,y="PetalLengthCm",x="Species")
```

```python
# setosa flowers have petal length of 1cm-2cm whereas versicolor have 3cm-5cm and virginica have 4.5cm-7cm.
# we can differentiate the flowers based on petal length.
```

###### 7. Create a Violin Plot.

```python
sns.violinplot(data=data,y="PetalLengthCm",x="Species",hue="Species")
```

```python
# more number of setosa flowers have petal length of appox 1.5cm
# more number of versicolor flowers have petal length of approx 4-5cm
# more number of virginica flowers have petal length of 5-6cm
```

##### 8. Create a pair chart.

```python
sns.pairplot(data,hue="Species",size=3)
```

```python
# We can differentiate the flowers based on petal length and petal width.
# If petal length is 1cm-2cm and petal width is 0-0.8cm all are setosa flowers,else petal length is between 3cm-5cm and petal width is between 1cm-2cm all are versicolor.
# If petal length is between 5cm-7cm and petal width is between 1.5cm -2.5cm all are virginica 
```
